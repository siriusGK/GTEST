```python
import pennylane as qml
import matplotlib.pyplot as plt
```

```python
dev=qml.device("default.qubit", wires=5) #making device available for 5 qubits
```

```python
def Hadamard_1():
    for i in range(5):
        qml.Hadamard(wires=i)#for superposition apply hadamard to all
        
```

```python
def cnot_1():
    qml.CNOT(wires=[0,1])
    qml.CNOT(wires=[1,2])
    qml.CNOT(wires=[2,3])   
    qml.CNOT(wires=[3,4])
```

```python
def swap_1():
    qml.SWAP(wires=[0,4])
```

```python
import numpy as np
def Rot_1():
    qml.RX(np.pi/2,wires=4)
```

```python
@qml.qnode(dev)#making decorater for functions
def circuit_Combined():
    Hadamard_1()
    cnot_1()
    swap_1()
    Rot_1()

    return qml.state()
```

```python
print(qml.draw(circuit_Combined)())
```
